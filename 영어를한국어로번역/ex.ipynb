```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
from IPython.display import clear_output 
import pandas as pd

d = pd.read_excel('refined_datafile.xlsx',index_col=0)
d2 = pd.read_excel('d2.xlsx', index_col=0)

prev = 0
for i in range(12360,len(d)):
    options = webdriver.ChromeOptions()
    options.add_argument("headless")
    driver = webdriver.Chrome('chromedriver.exe', chrome_options=options)
    driver.get('https://papago.naver.com/')
    driver.implicitly_wait(10)   
    try:
        english = driver.find_element_by_xpath('/html/body/div/div/div[1]/section/div/div[1]/div[1]/div/div[3]/label/textarea')
        text = d.iloc[i][1]    
        english.send_keys(text)
        clear_output()
        output_text = ' '
        element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "/html/body/div/div/div[1]/section/div/div[1]/div[2]/div/ul/li[1]/p")))
        # if len(text) > 1500:
        #     sleep(10)
        # else:
        #     sleep(5)
        
        korean = driver.find_element_by_xpath('/html/body/div/div/div[1]/section/div/div[1]/div[2]/div/div[5]/div/span')
        output_text = korean.text
    except:
        output_text = ' '
    
    if output_text == prev:
        output_text = ' '

    prev = output_text
    print(i,'=',output_text)

    # english.clear()
    driver.quit()
    
    d2.loc[i] = output_text
    if i % 30 == 0:
        d2.to_excel('d2.xlsx')


d2.to_excel('d2.xlsx')
   
d2
```
